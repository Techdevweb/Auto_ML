```python
import autosklearn
```

```python
#Regression 
import pandas as pd
df=pd.read_csv('housing.csv')
df
```

```python
target=df.pop('median_house_value')
df=pd.concat([df,target],axis=1) #Putting med_house at last
df.head()
```

```python
df.dtypes
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
```

```python
df['ocean_proximity']=le.fit_transform(df['ocean_proximity'])
df #Data replacing
```

```python
df=df.sample(frac=1,random_state=0)
df.reset_index(drop=True,inplace=True)
split_index=int(len(df)*0.75)
```

```python
train_df=df[:split_index]
test_df=df[split_index:]
```

```python
train_df
```

```python
X_train = train_df.to_numpy()[:, :-1]
y_train = train_df.to_numpy()[:, -1]

X_test = test_df.to_numpy()[:, :-1]
y_test = test_df.to_numpy()[:, -1]
#Converting data types to numpy

(X_train.shape, y_train.shape), (X_test.shape, y_test.shape)
```

```python
feature_types = ['Numerical']*8 + ['Categorical']

feature_types#Only for categorical encoding
```

```python
import autosklearn.regression

autosklearn_regressor = autosklearn.regression.AutoSklearnRegressor(
    time_left_for_this_task=240,
    per_run_time_limit=60,
)

autosklearn_regressor.fit(X_train, y_train, feat_type=feature_types)#Implementing auto-sklearn
```

```python
print(autosklearn_regressor.leaderboard())#Estimating best model
```

```python
from sklearn.metrics import mean_absolute_error as mae

mae(y_train, autosklearn_regressor.predict(X_train)), mae(y_test, autosklearn_regressor.predict(X_test))#Calculating mean square error
```

```python
result=autosklearn_regressor.predict(X_train)
result#Implementing model
```

```python
from sklearn.datasets import load_iris

iris = load_iris()

X2, y2 = iris.data, iris.target

X2.shape, y2.shape
```

```python
y2
```

```python
from sklearn.model_selection import train_test_split

X_train2, X_test2, y_train2, y_test2 = train_test_split(X2, y2, test_size=0.4, random_state=1)

X_train2.shape, y_train2.shape, X_test2.shape, y_test2.shape
```

```python
#Implementing auto-sklearn for classification
import autosklearn.classification

autosklearn_classifier = autosklearn.classification.AutoSklearnClassifier(
    time_left_for_this_task=80,
    per_run_time_limit=20,
)

autosklearn_classifier.fit(X_train2, y_train2) 
```

```python
print(autosklearn_classifier.leaderboard()) #Printing best model
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test2,autosklearn_classifier.predict(X_test2)))#Classification results
```
